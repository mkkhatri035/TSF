#### Prediction using Decision Tree Algorithm
With this we will create the Decision Tree classifier and visualize it graphically.

```python
#Importing modules required for classification
import pandas as pd
import numpy as np  

import matplotlib.pyplot as plt  
%matplotlib inline

from sklearn import tree
```

```python
iris=pd.read_csv("iris task6.csv")
iris.head()
```

```python
iris.drop('id',inplace=True,axis=1)
iris.tail()
```

```python
Species=iris['Species'].unique()
```

```python
len(iris)
```

```python
# import OneHotEncoder
from sklearn.preprocessing import OneHotEncoder,LabelEncoder
le=LabelEncoder()
iris['Species']=le.fit_transform(iris['Species'])
iris.head()
```

Creating dependent and independent variable

```python
X=iris.iloc[:,:-1].values
Y=iris.iloc[:,-1].values
print(X[0:5])
print(Y[0:5])
```

### Decision Tree Classifier Creation

```python
from sklearn.tree import DecisionTreeClassifier

classifier=DecisionTreeClassifier(criterion='entropy')
model=classifier.fit(X,Y)
```

### Decision Tree Classifier has been made and it could easily predict class for new data

## Visully representing Tree classifier


```python
fig = plt.figure(figsize=(25,20))
_=tree.plot_tree(model, 
                   feature_names=iris.columns[0:-1],
                   class_names=Species,
                   filled=True)
```

## Tree classifier is successfully created and represented Graphically
